```python
import pandas as pd
```

```python
bike_data = pd.read_csv("Data/out.csv")
bike_data.head()
```

```python
bike_data["starttime"]
```

```python
bike_data["hour"] = pd.to_datetime(bike_data["starttime"].values).hour
```

```python
bike_data.head()
```

```python
start_station = bike_data.groupby(["hour","start station name", "start station latitude", "start station longitude"]).count()["start station id"]
```

```python
start_station_df = pd.DataFrame(start_station).reset_index()
```

```python
start_station_df.head()
```

```python
start_station_df.to_csv("Data/citibike.csv")
```
